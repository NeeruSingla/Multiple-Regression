# Multiple Regression

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
boston_data=load_boston()
```

```python
df=pd.DataFrame(boston_data.data,columns=boston_data.feature_names)
df.head()
```

```python
df.shape
```

```python
X=df
```

```python
y=boston_data.target
```

## Statsmodels

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
X_constant=sm.add_constant(X)
```

```python
pd.DataFrame(X_constant)
```

```python
sm.OLS?
```

```python
model=sm.OLS(y,X_constant)
```

```python
lr=model.fit()
```

```python
lr.summary()
```

## statsmodels.formula.api

```python
form_lr=smf.ols(formula='y~ CRIM +  ZN + INDUS + CHAS + NOX + RM + AGE + DIS + RAD + TAX + PTRATIO + B + LSTAT',data=df)
mlr=form_lr.fit()
```

```python
mlr.summary()
```

## Detecting collinearity with Eigenvectors

```python
eigenvalues, eigenvectors =np.linalg.eig(df.corr())
pd.Series(eigenvalues).sort_values()
```

```python
np.abs(pd.Series(eigenvectors[:,8])).sort_values(ascending=False)
```

## Standardise Variable to Identify Key Feature(s)

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(X,y)
```

```python
result=pd.DataFrame(list(zip(model.coef_,df.columns)),columns=['coefficient','name']).set_index('name')
np.abs(result).sort_values(by='coefficient', ascending= False)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
scaler=StandardScaler()
Stand_coef_linear_reg=make_pipeline(scaler,model)
```

```python
Stand_coef_linear_reg.fit(X,y)
result=pd.DataFrame(list(zip(Stand_coef_linear_reg.steps[1][1].coef_,df.columns)),columns=['coefficient','name']).set_index('name')
np.abs(result).sort_values(by='coefficient', ascending=False)
```

## Use of R**2 to Identify Key Features

```python
from sklearn.metrics import r2_score
```

```python
linear_reg=smf.ols(formula='y~ CRIM +  ZN + INDUS + CHAS + NOX + RM + AGE + DIS + RAD + TAX + PTRATIO + B + LSTAT',data=df)
benchmark=linear_reg.fit()
r2_score(y, benchmark.predict(df))
```

without LSTAT

```python
linear_reg=smf.ols(formula='y~ CRIM +  ZN + INDUS + CHAS + NOX + RM + AGE + DIS + RAD + TAX + PTRATIO + B',data=df)
benchmark_without_LSTAT=linear_reg.fit()
r2_score(y, benchmark_without_LSTAT.predict(df))
```

without AGE

```python
linear_reg=smf.ols(formula='y~ CRIM +  ZN + INDUS + CHAS + NOX + RM + DIS + RAD + TAX + PTRATIO + B + LSTAT',data=df)
benchmark_without_age=linear_reg.fit()
r2_score(y, benchmark_without_age.predict(df))
```
